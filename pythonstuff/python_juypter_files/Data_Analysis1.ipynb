# Practice Problems

1. What was the highest scoring movie in 1996?

```python
#to answer this we want to do data[conditions where movie is from 1996 and score is max, relative to movies of 1996]
#alternatively we could select all movies from 1996, sort by score descending, and look at the top spot
data[data['Year'] == 1996].sort_values(by = 'Score', ascending = False)
#Fargo was the highest scoring movie, tied with The Bandit, in 1996
```

2. In what year was the highest rated movie of all time made?

```python
#To find this, we can just sort by score, and look at the year. 
data[['Year', 'Title', 'Score']].sort_values(by = 'Score', ascending = False)
#Shawshank and the Godfather are tied for the top. One is 1994 and the other 1972
```

3. What five movies have the most votes ever?

```python
#sort by votes, head(5)
data.sort_values(by = 'Votes', ascending = False).head(5)
#Shawshank, Dark Knight, Pulp Fiction, Godfather, Fight Club
```

4. What year in the 1960s had the highest average movie rating?

```python
#group by year, get average movie rating, select just in 1960-1969
sixties = data[(data['Year'] > 1959) & (data['Year'] < 1970)]

years = sixties.groupby(data['Year'])['Score'].mean()
years.sort_values(ascending = False)
#1962 has the highest average movie rating
```

# Cleaning Data

```python
#we want to remove the year from the title slot, genre into a list, and the "min" from length to int
#quickest way is to remove last 7 chars from title, or separate by ()
data['Title'] = data['Title'].str[:-7]
data
#note this method of editing in place does it again EVERY TIMe you run this
#data['Title'] = data['Title'].str.split('\(').str[0]
```

```python
data['Length'] = data['Length'].str.split().str.get(0).astype('int')
data
```
